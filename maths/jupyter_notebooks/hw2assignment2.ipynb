## Tehtävä 5.3.5
Määritä suorat, jotka kulkevat pisteiden
a) $(5, 0)$ ja $(0, 2)$
b) $(1, 2)$ ja $(3, -4)$
c) $(-5, 2)$ ja $(4, 7)$
kautta.

Suoran yhtälö on siis muotoa $y = kx + b$ jossa $k = \frac{y_2 - y_1}{x_2 - x_1}$ ja $b = y_1 - kx_1$.
Suoran yhtälöt laskin laiskana käyttäen [tätä laskinta](https://www.emathhelp.net/en/calculators/algebra-1/line-calculator/?lt=p&px=-5&py=2&qx=4&qy=7&s=&ppx=&ppy=).


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 10)
y1 =(-2/5) * x + 2
y2 = -3 * x + 5
y3 = (5/9) * x + (43/9)

ax = plt.subplot()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.plot(x, y1, label='a)')
plt.plot(x, y2, label='b)')
plt.plot(x, y3, label='c)')
plt.legend()
plt.show()
```

## 5.4.1
a) $y=x^2-4x+3$
b) $y=2x^2-40x+200$

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1, 12, 100)
y1 = x**2 - 4 * x + 3
y2 = 2*x**2 - 40 * x + 200

ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')

# x-akselin säätöä
#ax.set_xticks(np.arange(-1, 12, 1))


# Y:lle rajat, että kuvaajat näkyvät fiksummin
plt.ylim([-3, 7])

# label tarvitsee tuon legend():n
plt.plot(x, y1, label="$y=x^2-4x+3$")
plt.plot(x, y2, label="$y=2x^2-40x+200$")
plt.legend(loc=4)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 100, 100)
y = -0.0063 * x ** 2 + 0.55 * x

ax=plt.subplot()

ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# samat mittasuhteet
ax.set_aspect('equal')
ax.set_ylim([-1, 20])

plt.plot(x, y, color='blue', label='Flight path')
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.legend()

plt.show()
```
